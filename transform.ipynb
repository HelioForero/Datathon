```python
# Importamos las librerias

import pandas as pd
from sklearn.preprocessing  import OrdinalEncoder
import xgboost as xgb
import pickle

# Cargamos el dataset
df = pd.read_csv('.\hospitalizaciones_test.csv')

# Tomamos solo las columnas que nos intersan, deacuerdo a EDA.ipynb
df = df[['Department','gender','Ward_Facility_Code', 'Age', 'Available Extra Rooms in Hospital',
         'Admission_Deposit','Severity of Illness', 'doctor_name']] 

# Hacemos un ordinal encoder para transformar las columnas categoricas a numericas, igual que con el dataset de entrenamiento 
le = OrdinalEncoder()

df[['gender','Department','Ward_Facility_Code', 
'Severity of Illness', 'doctor_name','Age']] = le.fit_transform(df[['gender','Department','Ward_Facility_Code',
 'Severity of Illness', 'doctor_name','Age']])

df.head()
```

```python
# Traemos el modelo ya entrenado

filename = './xbg_trained.sav' 

loaded_model = pickle.load(open(filename, 'rb'))

# Predecimos las etiquetas
y_pred = loaded_model.predict(df)

y_pred = pd.DataFrame(y_pred, columns=['pred'])

# Guardamos las etiquetas para subirlas al datathon 
y_pred.to_csv("HelioForero.csv", index= False)
```
